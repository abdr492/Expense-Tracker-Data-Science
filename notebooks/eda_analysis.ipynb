```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
```

```python
df = pd.read_csv("../data/processed/cleaned_expenses.csv")
df["Date"] = pd.to_datetime(df["Date"])

df.head()
```

```python
expense_df = df[df["Type"] == "Expense"]

plt.figure(figsize=(8,5))
sns.barplot(x="Category", y="Amount", data=expense_df)
plt.title("Category-wise Expenses")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(expense_df["Amount"], bins=20, kde=True)
plt.title("Expense Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=expense_df["Amount"])
plt.title("Expense Outliers")
plt.show()
```

```python
df["Month"] = df["Date"].dt.month

monthly = df.groupby("Month")["Amount"].sum().reset_index()

plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Amount", data=monthly, marker="o")
plt.title("Monthly Expense Trend")
plt.show()
```
